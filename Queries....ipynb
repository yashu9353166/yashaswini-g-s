```python
def load_data(filename):
  crops=[]
  with open(filename,"r")as files:
   for line in files:
    line=line.strip()
    if line:
      data=line.split(",")
      crop={
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "rainfall":float(data[3]),
          "fertilizer":float(data[4]),
          "yield":float(data[5])
      }
      crops.append(crop)
      files.close()
      return crops
def calculated_yield_per_area(crop):
  return crop["yield"]/crop["area"]


def display_crops(crops):
  print("\n crop records")
  for crop in crops:
    ypa=calculated_yield_per_area(crop)
    print(crop["name"],"-",crop["season"],"yield/area:",round(ypa,2))

def find_highest_and_lowest_yield(crops):
  highest=crops[0]
  lowest=crops[0]
  for crop in crops:
    if calculated_yield_per_area(crop)>calculated_yield_per_area(highest):
      highest=crop
    if calculated_yield_per_area(crop)<calculated_yield_per_area(lowest):
      lowest=crop

  return highest,lowest


def seasonal_summary(crops):
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculated_yield_per_area(crop)
    if season not in summary:
      summary[season]=[]
      summary[season].append(ypa)
  return summary



```

```python
#crop counr per season
def crop_count_per_season(crops):
  count={}
  for crop in crops:
    season=crop["season"]
    if season not in count:
      count[season]=0
    count[season]+=1
  for season,value in count.items():
    print(season,":",value)
```

```python
#fertilizer efficicency analysis
def fertilizer_efficiency(crop):
  return crop["yield"]/crop["fertilizer"]
def most_fertilzer_efficient_crop(crops):
  best=crops[0]
  for crop in crops:
    if fertilizer_efficiency(crop)>fertilizer_efficiency(best):
      best=crop
  print("most fertilizer efficient crop:",best["name"])
```

```python
#rainfall efficiency analysis
def rainfall_efficiency_analysis(crops,max_rainfall):
  print("crop with rainfall <=",max_rainfall)
  for crop in crops:
    if crop["rainfall"]<=max_rainfall:
      print(crop["name"],"-",crop["yield"])
```

```python
#ranking crop based on productivity
def rank_crops_by_yield(crops):
  ranked=sorted(crops,key=calculated_yield_per_area,reverse=True)
  rank=1
  for crop in ranked:
    print(rank,crop["name"],"-",round(calculated_yield_per_area(crop),2))
    rank+=1
```

```python
#detection of crops grown in multiple season
def crops_in_multiple_seasons(crops):
 crop_seasons = {}
 for crop in crops:
  name = crop["name"]
  season = crop["season"]
  if name not in crop_seasons:
   crop_seasons[name] = []
  if season not in crop_seasons[name]:
   crop_seasons[name].append(season)


 for name, seasons in crop_seasons.items():
  if len(seasons) > 1:
    print(name, "grown in seasons:", seasons)
```

```python
def write_report(filename,crops,highest,lowest,summary):
  files=open(filename,"w")
  files.wrie("Agricultural Data Analysis report\n")
  files.write("crop yield per unit area:\n")
  for crop in crops:

    ypa=calculated_yield_per_area(crop)
    files.write(crop["name"]+":"+str(round(ypa,2))+"\n")
    files.write("\n Highest yield crop:\n")
    files.write(highest["name"]+"\n")
    files.write("\n lowest yield crop:\n")
    files.write(lowest["name"]+"\n")
    files.write("\n Seasonal average yield:\n")
    for season in summary:
      avg=sum(summary[season]/len(summary[season]))
      files.write(season+":"+str(round(avg,2))+"\n")

    files.close
```
