```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score

data={
    "rainfall":[450,520,390,610,480,530,410,590],
    "temperature":[22,20,25,19,23,21,24,18],
    "nitrogen":[0.8,0.1,0.6,1.2,0.9,1.1,0.7,1.3],
    "fertilizer":[120,140,100,160,130,150,110,170],
    "yield":[3.5,4.2,2.9,4.8,3.8,4.4,3.1,5.0]

}
df=pd.DataFrame(data)

x=df[["rainfall","temperature","nitrogen","fertilizer"]]
y=df["yield"]

print(x)
print(y)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)

print(x_train)
print(x_test)


model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

mae=mean_absolute_error(y_test,y_pred)

r2=r2_score(y_test,y_pred)

print("Mean Absolute Error:",round(mae,2))
print("R2 score:",round(r2,2))

new_data=np.array([[500,21,1.0,145]])
predicted_yield=model.predict(new_data)
print("Predicted Yield:",predicted_yield[0])

print("predicted yield (tons/hectare):",round(predicted_yield[0],2))
```

```python
def calculate_gdd(t_max, t_min, t_base=10):
    """
    Calculates GDD for a single day.
    t_base is usually 10°C for crops like corn.
    """
    avg_temp = (t_max + t_min) / 2
    gdd = max(avg_temp - t_base, 0)
    return gdd


print(f"GDD for today: {calculate_gdd(28, 16)}")
```

```python
def calculate_savings(principal, monthly_addition, annual_rate, years):
    monthly_rate = (annual_rate / 100) / 12
    months = years * 12
    total = principal

    for i in range(months):
        total = (total + monthly_addition) * (1 + monthly_rate)

    return total

# Try changing these numbers!
initial_deposit = 1000  # $1,000 to start
monthly_invest = 200    # $200 every month
rate = 7                # 7% average stock market return
years_to_wait = 20

final_balance = calculate_savings(initial_deposit, monthly_invest, rate, years_to_wait)
print(f"After {years_to_wait} years, you'll have: ${final_balance:,.2f}")
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)
n=120

rainfall=np.random.uniform(400,1200,n)
fertilizer=np.random.uniform(50,300,n)
temperture=np.random.uniform(18,35,n)

yield_tons=(0.005*rainfall+0.03*fertilizer-0.1*temperture+np.random.normal(0,2,n))

data=pd.DataFrame({
    "rainfall":rainfall,
    "fertilizer":fertilizer,
    "temperture":temperture,
    "yield":yield_tons
})

print(data.head())
```

```python
from sklearn.model_selection import train_test_split

x=data[["rainfall","fertilizer","temperture"]]
y=data["yield"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)

x_val,x_test,y_val,y_test=train_test_split(x_test,y_test,test_size=0.5,random_state=1)

print("Training set:",x_train.shape)
print("validation set.:",x_val.shape)
print("test set:",x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

def evaluate_model(x,y,name):
    y_pred=model.predict(x)
    mae=mean_absolute_error(y,y_pred)
    mse=mean_squared_error(y,y_pred)
    rmse=np.sqrt(mse)
    r2=r2_score(y,y_pred)


    print(f"\n{name}set performance")
    print("MAE:",mae)
    print("MSE:",mse)
    print("RMSE:",rmse)
    print("R2 score:",r2)

evaluate_model(x_train,y_train,"Training")
evaluate_model(x_val,y_val,"Validation")
evaluate_model(x_test,y_test,"Test")

```

```python
from sklearn.model_selection import cross_val_score

def cross_validation_results(k):
    scores=cross_val_score(model,x,y,scoring="neg_mean_squared_error",cv=k)
    rmse_scores=np.sqrt(-scores)

    print(f"\n{k}-fold cross validation results")
    print("RMSE scores:",rmse_scores)
    print("Mean RMSE:",np.mean(rmse_scores))
    print("Standard deviation:",np.std(rmse_scores))

for k in [3,5,10]:
    cross_validation_results(k)


```
