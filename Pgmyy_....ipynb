```python
student={
    "name":"anitha",
    "roll_no":101,
    "banch":"CSE",
    "marks":88
}
print("orginal dictionary:",student)



#copy()-crates a shallow copy
student_copy=student.copy()
print("copied  dictionary:",student_copy)
if student==student_copy:
  print("hi")
```

```python
#pop()-removes a specific key
removed_value=student.pop("name")
print("removed value :",removed_value)
print("after pop:",student)
```

```python
last_item=student.popitem()
print("removed item:",last_item)
print("after popitem:",student)
```

```python
#update()-adds or modify key-value pairs
student.update({"year":2,"branch":"AI & ML","batch":"a2","timing":"2 pm"})
print("after update:",student)
```
