```python
# integer variable

number_of_field=5
#float variable
area_per_field=2.5  #in acres
#string variable
crop_name="corn"
#boolean variable
irrigation_available=True
#display the variable
print("crop name:", crop_name)
print("number of field :", number_of_field)
print("area per field:", area_per_field,"acres")
print("irrigation available:", irrigation_available)
#using variable in calculation
total_area= number_of_field*area_per_field
print("\ntotal cultivated area:", total_area,"acres")
#simple decision using a variable
if irrigation_available:
    print("status:irrigation is available")
else:
    print("status:irrigation is not available")
```
