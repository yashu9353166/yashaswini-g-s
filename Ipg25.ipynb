```python
import pandas as pd
data={
    'farm_id':['F001','F002','F003','FOO4','F005','F006','F007'],
    'crop':['wheat','rice','corn','wheat','rice','corn','wheat'],
    'yield':[100,150,200,120,180,220,160],
    'rainfall':[500,600,700,550,650,700,800],
    'fertilizer':[120,150,200,110,160,195,125],
    'soil_ph':[6.5,5.8,6.2,6.7,5.9,6.1,6.4]
}

df=pd.DataFrame(data)
df['yield_per_ha']=(df['yield']/50).round(1)
print('original Dataset:')
print(df)
```

```python
print('\n===HIGH YIELD FARMS(>200 TONS)===')
high_yield=df[df['yield']>200]
print(high_yield)
```

```python
print('\n===RICE WITH HIGH RAINFALL (>800)===')
rice_high_rain=df[(df['crop']=='rice')&(df['rainfall']>800)]
print(rice_high_rain)
```

```python
print('\n===GRAIN CROPS (wheat & rice)===')
grain_crops=df[df['crop'].isin(['wheat','rice'])]
print(grain_crops[['farm_id','crop','yield']])
```

```python
print('\nb=== OPTIMAL SOIL PH(6.0 TO 6.5)===')
optimal_ph=df[df['soil_ph'].between(6.0 , 6.5)]
print(optimal_ph[['farm_id', 'soil_ph','yield']])
```

```python
print('\n=== COMPLEX QUERY:good condition===')
min_rainfall=600
good_conditions=df.query('rainfall>@min_rainfall and fertilizer>150 ')
print(good_conditions[['farm_id','crop','rainfall','fertilizer' ]])
```
